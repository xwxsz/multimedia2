# –£–ª—É—á—à–µ–Ω–Ω—ã–π –±–µ–π–∑–ª–∞–π–Ω

#### –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏

```python
model = YOLO("yolov8s.pt")

model.train(
    data="set/coco128.yaml",   
    epochs=20,                 
    imgsz=640,                 
    device="cpu",
    max_det=50,

    optimizer="SGD",
    lr0=0.01,                  
    lrf=0.01,                  
    momentum=0.937,
    weight_decay=0.0005,
    warmup_epochs=1,
    warmup_momentum=0.8,

    # –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø–æ—Ç–µ—Ä—å
    box=0.05,
    cls=0.5,
    dfl=1.0,

    # –†–∞—Å—à–∏—Ä–µ–Ω–Ω—ã–µ –∞—É–≥–º–µ–Ω—Ç–∞—Ü–∏–∏
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    degrees=0.3,
    translate=0.1,
    scale=0.9,
    shear=0.2,
    flipud=0.0,
    fliplr=0.5,
    mosaic=1.0,
    mixup=0.2,
    copy_paste=0.1,

    cos_lr=True,
    patience=10,              

    batch=8
)
```

#### –ú–µ—Ç—Ä–∏–∫–∏

```python
metrics = model.val()
print(f"mAP50: {metrics.box.map50:.4f}")
print(f"mAP50-95: {metrics.box.map:.4f}")
```

#### –í—ã–≤–æ–¥ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π

```python
image_dir = 'set/coco128/images/train2017'
image_files = random.sample(os.listdir(image_dir), 3)

for img_name in image_files:
    img_path = os.path.join(image_dir, img_name)
    results = model(img_path)
    img_plot = results[0].plot()
    
    plt.imshow(img_plot)
    plt.axis('off')
    plt.title(f"Prediction for {img_name}")
    plt.show()
```
